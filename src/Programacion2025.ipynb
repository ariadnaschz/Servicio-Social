<span style='font-size:large'>Creamos una función que nos ayude a eliminar datos duplicados, y principalmente arreglos en donde esta el mismo docente y materia, pero diferente valor numérico, esto con el objetivo con que se quede el arreglo con mayor peso con el fin de favorecer las opciones del docente, con este mismo propósito, también eliminamos los arreglos con valor numérico “cero”. </span>


```python
def filtrar(aristas):
    diccionario = {}
    for nombre, materia, valor in aristas:
        clave = (nombre.strip(), materia.strip())  # Eliminar espacios extra
        if clave in diccionario:
            diccionario[clave] = max(diccionario[clave], valor)
        else:
            diccionario[clave] = valor
    
    # Convertir de nuevo a lista y eliminar duplicados
    resultado = list({tuple([nombre, materia, valor]) for (nombre, materia), valor in diccionario.items()})
    
    # Eliminar arreglos con valor 0
    resultado = [arista for arista in resultado if arista[2] > 0]
    
    return resultado

Aristas = filtrar(Aristas25I)
len(Aristas)
```

<span style='font-size:large'>A partir de estos datos generamos las gráficas asociadas donde los vértices son los profesores y las UEAs, hay una arista entre un profesor y una UEA si el profesor puso esa UEA en su lista de preferencias. Así mismo las aristas estarán ponderadas con los valores de la columna pesos.</span>

<span style='font-size:large'>Entonces creamos una gráfica bipartita para analizar la relación entre docente y UEA. De esta forma optimizamos las asignaciones mediante un emparejamiento máximo.</span>
