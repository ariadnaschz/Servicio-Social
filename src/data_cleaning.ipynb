## Exportación de los datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("../data/Data.xlsx")
df.info()
```

```python
# Se seleccionan solo las columnas con las que trabajaremos
df = df[['Nombre Completo',
         'Selecciona tu primera opción, es decir, la opción que más te agradaría impartir.',
         'Selecciona tu segunda opción.',
         'Selecciona tu tercera opción.',
         'Selecciona tu cuarta opción.',
         'Selecciona tu quinta opción.',
         'Selecciona tu primera opción, es decir la opción que más te agradaría impartir.',
         'Selecciona tu segunda opción. 2', 
         'Selecciona tu tercera opción. 2',
         'Selecciona tu cuarta opción. 2', 
         'Selecciona tu quinta opción. 2',
         'Selecciona tu primera opción, es decir la opción que más te agradaría impartir. 2',
         'Selecciona tu segunda opción. 3', 
         'Selecciona tu tercera opción. 3',
         'Selecciona tu cuarta opción. 3', 
         'Selecciona tu quinta opción. 3',
         ]]

df.info()
```

```python
# Diccionario con los nuevos nombres para las columnas
nuevos_nombres = {
    0: "Docente",
    1: "1_opcion_25I",
    2: "2_opcion_25I",
    3: "3_opcion_25I",
    4: "4_opcion_25I",
    5: "5_opcion_25I",
    6: "1_opcion_25P",
    7: "2_opcion_25P",
    8: "3_opcion_25P",
    9: "4_opcion_25P",
    10: "5_opcion_25P",
    11: "1_opcion_25O",
    12: "2_opcion_25O",
    13: "3_opcion_25O",
    14: "4_opcion_25O",
    15: "5_opcion_25O"
}

# Renombrar columnas
df.rename(columns={df.columns[i]: nombre for i, nombre in nuevos_nombres.items()}, inplace=True)
df.head()
```

```python
# Diccionario de pesos por número de opción
pesos = {
    "1": 190,
    "2": 94,
    "3": 46,
    "4": 22,
    "5": 10
}

# Lista de prefijos por periodo
periodos = ['25I', '25P', '25O']

# Recorrer cada combinación de opción y periodo
for i in range(1, 6):  # opciones 1 a 5
    for p in periodos:
        nombre_columna = f"{i}_opcion_{p}"
        nombre_peso = f"Peso{i}_{p}"

        # Crear nueva columna de peso con el valor correspondiente
        if nombre_columna in df.columns:
            df[nombre_peso] = pesos[str(i)]

pd.set_option('display.max_columns', None)
df.head()
```

```python
# Índices de fila donde se quiere cambiar el valor para docentes específicos
indices_objetivo = [0, #Docente 1
                    1, #Docente 2
                    3, #Docente 4
                    15 #Docente 16
                    ]

# Columnas a modificar
columnas_peso = ["Peso1_25I", "Peso1_25P", "Peso1_25O"]

# Cambiar el valor a 20000 en las posiciones indicadas
for idx in indices_objetivo:
    for col in columnas_peso:
        df.at[idx, col] = 20000

df.loc[indices_objetivo, ["Docente"] + columnas_peso]
```

## Arreglo para las aristas del trimestre 25I

```python
# Lista donde guardaremos los resultados del trimestre 25I
Aristas25I = []

# Recorremos cada fila del DataFrame
for _, fila in df.iterrows():
    for i in range(1, 6):  # Opciones 1 a 5
        nombre_opcion_25I = f"{i}_opcion_25I"
        nombre_peso_25I = f"Peso{i}_25I"

        # Verificamos que existan las columnas antes de acceder
        if nombre_opcion_25I in df.columns and nombre_peso_25I in df.columns:
            Aristas25I.append([
                fila["Docente"],
                fila[nombre_opcion_25I],
                fila[nombre_peso_25I]
            ])

for fila in Aristas25I[:5]:
    print(fila)
```

```python
# UEA que deben duplicarse ya que se imparten en dos grupos
ueas_a_duplicar = ["UEA 4", "UEA 3", "UEA 2", "UEA 18", "UEA 19"]

# Lista para guardar los duplicados
duplicados = []

# Recorrer una copia del arreglo original para no modificarlo mientras iteramos
for i, fila in enumerate(Aristas25I):
    docente, uea, peso = fila
    if uea in ueas_a_duplicar:
        # Cambiar el nombre original a "uea I1"
        Aristas25I[i][1] = f"{uea} I1"

        # Crear duplicado con nombre "uea I2"
        duplicado = [docente, f"{uea} I2", peso]
        duplicados.append(duplicado)

# Agregar los duplicados al arreglo original
Aristas25I.extend(duplicados)

# Mostrar últimos resultados para ver los cambios
for fila in Aristas25I[-10:]:
    print(fila)
```

```python
# Eliminar subarreglos con algún valor nulo
Aristas25I = [
    fila for fila in Aristas25I
    if all(valor is not None and valor == valor for valor in fila)
]
```

```python
Aristas25I
```

Guardamos los arreglos en un archivo txt para que poder compiarlos mejor en el programa de asignación 

```python
with open('Aristas25I.txt', 'w') as f:
  f.write(str(Aristas25I))
```
